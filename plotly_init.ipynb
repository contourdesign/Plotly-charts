```python
import plotly
from plotly import __version__
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
# Allow us to display plots in Jupyter notebook
init_notebook_mode(connected=True)
```

```python
# example plot
iplot([{"x": [1,2,3,4], "y": [1,2,6,4]}])
```

```python
# Heatmap contour
import plotly.graph_objs as go
import numpy as np

x = np.random.randn(2000)
y = np.random.randn(2000)
iplot([go.Histogram2dContour(x=x, y=y, contours=dict(coloring="heatmap")),
      go.Scatter(x=x, y=y, mode='markers', marker=dict(color='white', size=3, opacity=0.3,))], show_link=False)
```

# Simple Scatterplot

```python
# Simple Scatterplot
N = 500
example_x = np.random.rand(N)
example_y = np.random.rand(N)
```

```python
trace = go.Scatter(
    x=example_x,
    y=example_y,
    mode='markers')
```

```python
data = [trace]
iplot(data)
```

# Simple Pie

```python
groups = ['Rest', 'Food', 'Bills', 'Misc']

amount = [1000, 518, 331, 277]

colors = ['#d32e58', '#f9blee', '#b7f9b1', '#b1f5f9']

trace = go.Pie(labels=groups, values=amount, hoverinfo='label+percent', textinfo='value',
              textfont=dict(size=25),
              marker=dict(colors=colors, line=dict(color='#000000', width=3)))

iplot([trace])
```
