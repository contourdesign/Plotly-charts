# Mapping in plotly

```python
import plotly.figure_factory as ff
import pandas as pd

import pandas as pd
import numpy as np
import plotly
from plotly import __version__
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```python
df = pd.read_csv('minority-majority.csv')
df.head()
```

```python
scope = ['California']

df_new = df[df['STNAME'].isin(scope)]

values = df_new['TOT_POP'].tolist()

fips = df_new['FIPS'].tolist()
```

```python
# generate random colors for colorscale
import random
get_colors = lambda n: list(map(lambda i: "#" + "%06x" % random.randint(0, 0xFFFFFF),range(n)))
colorscale = get_colors(100) 
```

```python

fig = ff.create_choropleth(
    fips=fips, values=values, scope=scope, colorscale=colorscale,
    round_legend_values=True, simplify_county=0, simplify_state=0,
    county_outline={'color': 'rgb(15,15,55)', 'width': 0.5},
    state_outline={'width': 0.5},
    legend_title='Population per county',
    title='California')

iplot(fig, filename='choropleth')
```
